```python
#Verificar nulos
dados_nan = meus_dados[meus_dados.isna().any(axis=1)]
dados_nan.shape
dados_nan.head(50)
```

```python
#Convertendo String para Float com duas casas decimais

# Crie uma cópia do DataFrame original
meus_dados_copia = meus_dados.copy()

# Lista das colunas que você deseja converter
colunas_para_converter = ['Salario', 'FGTS', '13Sal', 'Ferias']

# Loop através das colunas e substitua vírgulas por pontos
for coluna in colunas_para_converter:
    meus_dados_copia[coluna] = meus_dados_copia[coluna].str.replace(',', '.', regex=True)

# Converter as colunas para float
meus_dados_copia[colunas_para_converter] = meus_dados_copia[colunas_para_converter].astype(float)

# Arredonde as colunas de tipo float para duas casas decimais
colunas_float = meus_dados_copia.select_dtypes(include=['float'])
meus_dados_copia[colunas_float.columns] = colunas_float.round(2)
```

```python
#Identificando caracteres fora do padrão

meus_dados_copia2 = meus_dados_copia.copy()

# Especifique a coluna em que deseja verificar os caracteres fora do padrão
colunas_verificar = ['Nome', 'Nascimento', 'Filhos', 'Ativo', 'UF', 'email', 'CestaBasica','AjudaCusto']


# Use uma expressão regular para identificar caracteres fora do padrão (letras e números)
padrao = r'[^a-zA-Z0-9]'  # Esta expressão regular corresponde a qualquer coisa que não seja letra ou número

registros_fora_do_padrao = pd.DataFrame()  # DataFrame vazio para armazenar os registros fora do padrão
for coluna in colunas_verificar:
    mascara = meus_dados_copia2[coluna].str.contains(padrao, regex=True, na=False)
    registros_com_problema = meus_dados_copia2[mascara]
    registros_fora_do_padrao = pd.concat([registros_fora_do_padrao, registros_com_problema])
print(registros_fora_do_padrao)
```

```python
# Tratando os caracteres fora do padrão
substituicoes = {
    'Æ': 'ã',
    '‡': 'ç',
    '£': 'ú'
}

for coluna in meus_dados_copia2.columns:
    if meus_dados_copia2[coluna].dtype == 'object':  # Verifique se a coluna contém dados de objeto (string)
        for antigo, novo in substituicoes.items():
            meus_dados_copia2[coluna] = meus_dados_copia2[coluna].str.replace(antigo, novo)
meus_dados_copia2.head()
```

```python
# Especifique o caminho do arquivo Excel de saída
caminho_arquivo_excel = 'meus_dados_copia2.xlsx'

# Use o método to_excel() para exportar o DataFrame para um arquivo Excel
meus_dados_copia2.to_excel(caminho_arquivo_excel, index=False)
```
